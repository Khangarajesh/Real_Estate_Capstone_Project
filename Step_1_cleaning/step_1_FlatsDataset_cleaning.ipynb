#Dealing with price column

```python
flats['price'].head(400).unique()
```

We can see the price column is not consistent. It contains values in crore, lakh and Price on Request.we need to make thiis column consistent.




```python
flats[flats['price'] == 'Price on Request'].shape[0]
```

Since this column is going to become our target colum and we can not keep the values like 'Price on Request'. There are only 11 such a rows so we will drop this rows

```python
flats = flats[flats['price'] != 'Price on Request']
```

```python
#create a function to keep unit of all values in price column in Crore
```

```python
def price_in_crore_KHANGAR(x):
  if type(x) == float:
    return x
  elif x[1] == 'Lac':
    return round(float(x[0])/100, 2)
  else:
    return round(float(x[0]), 2)

```

```python
flats.drop(286, inplace = True) #row containing value 'price'
```

```python
flats['price'] = flats['price'].str.split(' ').apply(lambda x: price_in_crore_KHANGAR(x))
```

```python
flats.sample(2)
```

#Dealing with price_per_sqft column

```python
flats['price_per_sqft'].str.split('/').str[1].unique()
```

we can see price_per_sqft column is consistent so we will just remove the unnecessary part

```python
flats['price_per_sqft'] = flats['price_per_sqft'].str.split('/').str[0].str.replace('₹', '').str.replace(',', '').astype(float)
```
