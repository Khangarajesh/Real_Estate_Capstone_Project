#sector and price

```python
import re
```

```python
sector_avg_price = df.groupby('sector')['price'].mean().reset_index()
```

```python
def extract_num_from_sec(sector_name):
  pattern = r'\b\d+\b'
  matches = re.findall(pattern, sector_name)
  if matches:
    return int(matches[0])
  else:
    return float('inf')
```

```python
sector_avg_price['sec_num'] = sector_avg_price['sector'].apply(lambda x : extract_num_from_sec(x))
```

```python
sector_avg_price = sector_avg_price.sort_values(by = 'sec_num')
```

#price_per_sqft

```python
#plt.figure(figsize = (5,50))
#sns.heatmap(sector_avg_price.set_index('sector')[['price']], annot =True, fmt = ".2f", linewidth = .5)
```

```python
sector_price_per_sqft = df.groupby('sector')['price_per_sqft'].mean().reset_index()
```

```python
sector_price_per_sqft['sec_num'] = sector_price_per_sqft['sector'].apply(extract_num_from_sec)
```

```python
sector_price_per_sqft = sector_price_per_sqft.sort_values(by = 'sec_num')
```

#luuxary_score

```python
sector_luxary_score = df.groupby('sector')['luxary_score'].mean().reset_index()
```

```python
sector_luxary_score['sec_num'] = sector_luxary_score['sector'].apply(extract_num_from_sec)
sector_luxary_score = sector_luxary_score.sort_values(by = 'sec_num')
```

```python
plt.figure(figsize = (10,80))
plt.subplot(1,3,1)
sns.heatmap(sector_price_per_sqft.set_index('sector')[['price_per_sqft']], annot = True, fmt = ".2f", linewidth = .5)

plt.subplot(1,3,2)
sns.heatmap(sector_avg_price.set_index('sector')[['price']], annot =True, fmt = ".2f", linewidth = .5)

plt.subplot(1,3,3)
sns.heatmap(sector_luxary_score.set_index('sector')[['luxary_score']], annot =True, fmt = ".2f", linewidth = .5)
```

##obsrvation

* At some places price_per_sqft is higher but price is lower which indicated the average area in those particular sector is less but very expensive

* At some places I observe price is higher but price_per_sqft is less which indicates the average area in those places is higher but less expensive
